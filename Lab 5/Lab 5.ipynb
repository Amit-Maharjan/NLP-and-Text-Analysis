**Q1.1.** Preprocess the `text` field content by performing the necessary steps to make it ready for applying sentiment analysis. Append the resulted preprocessed text to the `reviews_df` as a new column called `preprocessed_review`.

```python
reviews_df = reviews_df.drop_duplicates(subset=['ProductId', 'UserId', 'Text'])
```

```python
reviews_df.shape
```

```python
import nltk
nltk.download('wordnet')
```

```python
from nltk.stem import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()

def preprocess(documents):
    processed_docs = []
    for doc in documents:
        tokens = doc.lower().split()
        lemmatized_tokens = [lemmatizer.lemmatize(word) for word in tokens]
        processed_docs.append(" ".join(lemmatized_tokens))
    return processed_docs

reviews_df['preprocessed_review'] = preprocess(reviews_df['Text'])
```

```python
reviews_df.head()
```

**Q1.2.** Using nltk's `SentimentIntensityAnalyzer` module, get the `neg`, `neu`, `pos`, and `compound` scores of the `preprocessed_review` field and append the result to the `reviews_df` as new 4 columns-`neg`, `neu`, `pos`, and `compound`, respectively.

```python
nltk.download('vader_lexicon')
```

```python
from nltk.sentiment import SentimentIntensityAnalyzer

sia = SentimentIntensityAnalyzer()
```

```python
res = {}
for i, row in reviews_df.iterrows():
    text = row['preprocessed_review']
    myid = row['Id']
    res[myid] = sia.polarity_scores(text)
```

```python
res = pd.DataFrame(res).T
res = res.reset_index().rename(columns={'index': 'Id'})

res['Id'] = res['Id'].astype(str)
reviews_df['Id'] = reviews_df['Id'].astype(str)

reviews_df = res.merge(reviews_df, how='left', on='Id')
```

```python
reviews_df.head()
```

**Q1.3.** Label each review in the `reviews_df` as either `pos` or `neg` based on the `compound` value for that specific review, and append the result to the dataframe as new column called `comp_score`. If the value of `compound` is greater than or equals to `zero`, then the `comp_score = 'pos'` and `'neg'` otherwise.

```python
reviews_df['comp_score'] = reviews_df['compound'].apply(lambda x: 'pos' if x >= 0 else 'neg')
```

**Q1.5.** Based on the plots generated in `Q1.4`, discuss the results of the analysis.

**Answer:** The bar plot of Negative Score vs Actual Score is positively skewed. This indicates that users tend to provide more negative reviews for products with lower actual scores. Non-technically speaking, this makes sense: if a user has a poor experience with a product, they are likely to give it a low score and leave negative comments. 

As the actual score increases, the negative score decreases. However, we can also observe a few negative comments (as indicated by the negative score) for higher actual scores, such as 5. This might be due to incorrect user input, like a user entering a rating of 5 but leaving a negative comment, or it could indicate that our model is not functioning properly.

Similarly, for Positive Score vs Actual Score and Compound Score vs Actual Score (which combines negative and positive scores), we can observe negative skew. This also makes sense because, as the rating increases, positive reviews tend to increase as well.

However, the graph for Neutral Score vs Actual Score is constant. Across all actual scores, neutral comments remain almost consistent. This suggests that neutral comments are evenly distributed across all rating categories.
