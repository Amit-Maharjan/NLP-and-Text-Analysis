**Q1.1.** Perform Part-of-Speech tagging on the `Sentence` column in the `ner_df`. Then, add the results as new column called `pos`.

```python
# Q1.1. Perform Part-of-Speech tagging on the Sentence column in the ner_df. Then, add the results as new column called pos.
pos_df = []
for sentence in ner_df['Sentence']:
    tokens = nltk.word_tokenize(sentence)
    tags = nltk.pos_tag(tokens)
    pos_df.append(tags)

ner_df['pos'] = pos_df
```

```python
ner_df.head(10)
```

```python
pprint(ner_df['pos'][0])
```

**Q1.2.** Implement a function called `np_chunker` that receives each sentence from the `ner_df` dataframe. Then, it defines and applies a chunk parser to chunk all noun phrases from these sentences.

```python
# Q1.2. Implement a function called np_chunker that receives each sentence from the ner_df dataframe. Then, it defines and applies a chunk parser to chunk all noun phrases from these sentences.
grammar = "NP: {<DT>?<JJ.*>*<NN.*>+}"
cp = nltk.RegexpParser(grammar)
def np_chunker(sentence):
    tokens = nltk.word_tokenize(sentence)
    pos_tags = nltk.pos_tag(tokens)
    result = cp.parse(pos_tags)

    noun_phrases = []
    for subtree in result.subtrees(filter=lambda t: t.label() == 'NP'):
        np = " ".join([token for token, pos in subtree. leaves()])
        noun_phrases.append(np)

    return noun_phrases
    
ner_df['Noun Phrases'] = ner_df['Sentence'].apply(np_chunker)
```

```python
ner_df.head(10)
```

```python
pprint(ner_df['Noun Phrases'][0])
```

**Q1.3.** Perfrom a Named Entity Recognition (NER) on the Sentence column and add the extracted entities as a new column called `entities`

```python
# Q1.3. Perfrom a Named Entity Recognition (NER) on the Sentence column and add the extracted entities as a new column called entities
def namedEntityRecognition(sentence):
    tokens = nltk.word_tokenize(sentence)
    pos_tags = nltk.pos_tag(tokens)
    chunks = nltk.ne_chunk(pos_tags)
    
    entities = []
    for chunk in chunks:
        if hasattr(chunk, 'label'):
            entity = " ".join([word for word, pos in chunk.leaves()])
            entities.append((chunk.label(), entity))

    return entities

# Function to apply NER in batches
def apply_ner_in_batches(df, batch_size):
    for start in range(0, len(df), batch_size):
        end = start + batch_size
        df.loc[start:end-1, 'entities'] = df.loc[start:end-1, 'Sentence'].apply(namedEntityRecognition)

apply_ner_in_batches(ner_df, 1000)
```
