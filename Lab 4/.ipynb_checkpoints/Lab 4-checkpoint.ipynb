**Q1.1.** Preprocess the `headline_text` field content by performing the necessary steps to make it ready for applying topic modeling algorithms. Store the resulted preprocessed text into a list called `headlines`

```python
df = df.drop_duplicates(subset='headline_text', keep='first')
```

```python
df.shape
```

```python
df.head(10)
```

```python
from nltk.corpus import stopwords
stop_words = stopwords.words('english')

from nltk.stem import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()

def preprocess(documents):
    processed_docs = []
    for doc in documents:
        tokens = doc.lower().split()
        filtered_tokens = [token for token in tokens if token not in stop_words and token.isalpha()]
        lemmatized_tokens = [lemmatizer.lemmatize(word) for word in filtered_tokens]
        processed_docs.append(" ".join(lemmatized_tokens))
    return processed_docs

headlines = preprocess(df['headline_text'])
```

```python
from pprint import pprint 
pprint(headlines[:10])
```

**Q1.2.** Using `Explicit Symantic Analysis` algorithm, find and print the top `20` related documents to the following document: ['price', 'sugar', 'sweet'\]. 

```python
from sklearn.feature_extraction.text import TfidfVectorizer

tfidf_vectorizer = TfidfVectorizer()
tfidf_matrix = tfidf_vectorizer.fit_transform(headlines)
print("TF-IDF Matrix shape:", tfidf_matrix.shape)
```

```python
searched_document = ['price', 'sugar', 'sweet']
searched_document_tfidf = tfidf_vectorizer.transform([' '.join(searched_document)])
```

```python
from sklearn.metrics.pairwise import cosine_similarity
cosine_similarities = cosine_similarity(searched_document_tfidf, tfidf_matrix).flatten()
```

```python
top_20_indices = cosine_similarities.argsort()[-20:][::-1]
```

```python
print("Top 20 related documents:")
for i in top_20_indices:
    print(f"{headlines[i]} - {cosine_similarities[i]:.4f}")
```

**Q1.3.** Using only the first `50` headlines, perform a document clustering using `AgglomerativeClustering` algorithm and plot the resulted dendrogram.

```python
from sklearn.cluster import AgglomerativeClustering 
model = AgglomerativeClustering(distance_threshold=0, n_clusters=None) 
model = model.fit(tfidf_matrix[:50].toarray())
```

```python
model.n_features_in_
```

```python
def plot_dendrogram(model, **kwargs) :
    # Create linkage matrix and then plot the dendrogram
    
    # Create the counts of samples under each node
    counts = np.zeros(model.children_.shape[0]) 
    n_samples = len(model.labels_)
    for i, merge in enumerate (model.children_):
        current_count = 0 
        for child_idx in merge: 
            if child_idx < n_samples:
                current_count += 1 # leaf node 
            else:
                current_count += counts[child_idx - n_samples] 
            counts[i] = current_count
            
    linkage_matrix = np.column_stack([model.children_, model.distances_, counts]).astype(float)
    # Plot the corresponding dendrogram 
    dendrogram(linkage_matrix, **kwargs)
```

```python
import numpy as np 
import matplotlib.pyplot as plt
from scipy.cluster.hierarchy import dendrogram 

plt.figure(figsize=(16,6))
plot_dendrogram(model)
```

**Q1.4.** Using `Latent Dirichlett Allocation` algorithm, perform topic modeling to find `20` topics from the headlines. Then, print the top `10` words for each topic. 

```python
import gensim
import gensim.corpora as corpora

id2word = corpora.Dictionary([doc.split() for doc in headlines])
corpus = [id2word.doc2bow(text.split()) for text in headlines]

lda_gensim_model = gensim.models.LdaModel(corpus, num_topics=20, id2word=id2word, passes=10)
```

```python
topics = lda_gensim_model.print_topics(num_words=10)
for idx, topic in topics:
    print(f"Topic #{idx}: {topic}")
    print("\n")
```

```python
from collections import defaultdict

# Assign each document to its most dominant topic
document_topics = defaultdict(list)
for idx, doc in enumerate(corpus):
    topic_distribution = lda_gensim_model.get_document_topics(doc)
    # Find the topic with the highest probability for this document
    dominant_topic = max(topic_distribution, key=lambda x: x[1])[0]
    document_topics[dominant_topic].append(idx)

# Display sample documents for each topic
num_samples = 4  # Number of documents to display per topic
print("\nSample Documents for Each Topic Using Gensim LDA:")

for topic_id, doc_indices in sorted(document_topics.items()):
    print(f"\nTopic {topic_id + 1}:")
    print(f"Number of documents assigned to this topic: {len(doc_indices)}")
    print(f"Displaying {min(num_samples, len(doc_indices))} sample documents:\n")

    for i in range(min(num_samples, len(doc_indices))):
        doc_index = doc_indices[i]
        print(f"Sample Document {i + 1} (Document ID: {doc_index}):")
        # Display the first 100 words from the document
        print(' '.join(headlines[doc_index].split()[:100]) + "...\n")
```

```python
import pyLDAvis
import pyLDAvis.gensim_models as gensimvis

# Visualize the topics
pyLDAvis.enable_notebook()
lda_display = gensimvis.prepare(lda_gensim_model, corpus, id2word, sort_topics=False)
pyLDAvis.display(lda_display)
```

**Q1.5.** Using `Non-Negative Matrix Factorization` algorithm, perform topic modeling to find `20` topics from the headlines. Then, print the top `10` words for each topic.

```python
from sklearn.decomposition import NMF
nmf_model = NMF(n_components=20, random_state=42)
nmf_topic_matrix = nmf_model.fit_transform(tfidf_matrix)

print("NMF Topic Matrix shape:", nmf_topic_matrix.shape)
```

```python
# Extract the top 10 words for each topic
num_top_words = 10
feature_names = tfidf_vectorizer.get_feature_names_out()

for topic_idx, topic in enumerate(nmf_model.components_):
    print(f"Topic {topic_idx + 1}:")
    top_words = [feature_names[i] for i in topic.argsort()[:-num_top_words - 1:-1]]
    print(", ".join(top_words))
```

```python
# Number of sample documents to display per topic
num_samples = 4

# Find the top documents for each topic
for topic_idx in range(nmf_model.n_components):
    print(f"Documents most associated with Topic {topic_idx + 1}:")

    # Get the column corresponding to the current topic
    topic_column = nmf_topic_matrix[:, topic_idx]

    # Get indices of the documents sorted by their importance to the topic
    top_doc_indices = np.argsort(topic_column)[::-1][:num_samples]

    # Display sample documents
    for doc_num, doc_index in enumerate(top_doc_indices):
        print(f"\nSample document {doc_num + 1} (Document ID: {doc_index}):")
        # Display the first 100 words from the document
        print(''.join(headlines[doc_index][:100]) + "...")
    print('--------------------------------------')
```
