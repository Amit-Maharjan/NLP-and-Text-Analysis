```python
import numpy as np
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer, CountVectorizer
from sklearn.decomposition import TruncatedSVD, NMF
from sklearn.cluster import KMeans
from sklearn.decomposition import LatentDirichletAllocation
import gensim
import gensim.corpora as corpora
import nltk
from nltk.corpus import stopwords

import matplotlib.pyplot as plt
stop_words = stopwords.words('english')

import warnings

# Suppress the specific deprecation warning
warnings.filterwarnings("ignore", category=DeprecationWarning)
```

## Load 20 Newsgroups Dataset
We’ll fetch the "20 Newsgroups" dataset using scikit-learn’s built-in utility. We select three categories: rec.sport.hockey, sci.space, and comp.graphics to keep it manageable for demonstration.

For the most accurate and updated information on the "20 Newsgroups" dataset, refer to the [official scikit-learn documentation](https://scikit-learn.org/stable/modules/generated/sklearn.datasets.fetch_20newsgroups.html). 

```python
from sklearn.datasets import fetch_20newsgroups
```

```python
# Fetch the dataset
newsgroups = fetch_20newsgroups(subset='all', categories=['rec.sport.hockey', 'sci.space', 'comp.graphics'], remove=('headers', 'footers', 'quotes'))
documents = newsgroups.data

# Display a sample
print("Sample document:\n", documents[0])
```

## Preprocess the Dataset

This step removes unnecessary characters and stopwords, leaving only meaningful content.

```python
def preprocess(documents):
    processed_docs = []
    for doc in documents:
        tokens = doc.lower().split()
        filtered_tokens = [token for token in tokens if token not in stop_words and token.isalpha()]
        processed_docs.append(" ".join(filtered_tokens))
    return processed_docs

processed_docs = preprocess(documents)
print("Sample preprocessed document:\n", processed_docs[0])
```

## 1. Explicit Semantic Analysis (ESA)

We limit the TF-IDF matrix to 1,000 features for simplicity and performance.

```python
tfidf_vectorizer = TfidfVectorizer(max_features=1000)  # Limiting to 1000 features for simplicity
tfidf_matrix = tfidf_vectorizer.fit_transform(processed_docs)

print("TF-IDF Matrix shape:", tfidf_matrix.shape)
```

## 2. Document Clustering Using KMeans

Here, num_clusters is set to 3, matching the categories in our subset.

```python
num_clusters = 3
km_model = KMeans(n_clusters=num_clusters, random_state=42)
km_model.fit(tfidf_matrix)

clusters = km_model.labels_.tolist()
print("Sample Cluster Assignments:", clusters[:10])
```

```python
from collections import defaultdict

# Group documents by their assigned cluster
clustered_docs = defaultdict(list)
for idx, cluster in enumerate(clusters):
    clustered_docs[cluster].append(processed_docs[idx])

# Display four sample documents for each cluster
num_samples = 4  # Number of documents to display per cluster
for cluster_id, docs in clustered_docs.items():
    print(f"\nCluster {cluster_id} (Topic {cluster_id + 1}):")
    print(f"Number of documents in this cluster: {len(docs)}")
    print(f"Displaying {min(num_samples, len(docs))} sample documents:\n")

    for i in range(min(num_samples, len(docs))):
        print(f"Sample document {i + 1}:\n{''.join(docs[i][:100])}...\n")  # Displaying the first 100 words
```

## 3. Latent Semantic Analysis (LSA)

We reduce the TF-IDF matrix to a 3-topic representation using SVD.

```python
lsa_model = TruncatedSVD(n_components=3, random_state=42)
lsa_topic_matrix = lsa_model.fit_transform(tfidf_matrix)

print("LSA Topic Matrix shape:", lsa_topic_matrix.shape)
```

```python
# Number of sample documents to display per topic
num_samples = 4

# Find the top documents for each topic
for topic_idx in range(lsa_model.n_components):
    print(f"\nDocuments most associated with Topic {topic_idx + 1}:")

    # Get the column corresponding to the current topic
    topic_column = lsa_topic_matrix[:, topic_idx]

    # Get indices of the documents sorted by their importance to the topic
    top_doc_indices = np.argsort(topic_column)[::-1][:num_samples]

    # Display sample documents
    for doc_num, doc_index in enumerate(top_doc_indices):
        print(f"\nSample document {doc_num + 1} (Document ID: {doc_index}):")
        # Display the first 100 words from the document
        print(''.join(processed_docs[doc_index][:100]) + "...")
```

#### Visualize LSA Topics:

```python
plt.scatter(lsa_topic_matrix[:, 0], lsa_topic_matrix[:, 1], c=clusters)
plt.title('LSA Topic Plot (3 categories)')
plt.xlabel('Topic 1')
plt.ylabel('Topic 2')
plt.show()
```

## 4. Latent Dirichlet Allocation (LDA)

Both of the following implementations show LDA's ability to identify three distinct topics.
##### Using Scikit-learn

```python
lda_model = LatentDirichletAllocation(n_components=3, random_state=42)
lda_topic_matrix = lda_model.fit_transform(tfidf_matrix)

print("LDA Topic Matrix shape:", lda_topic_matrix.shape)
```

##### Using Gensim

```python
id2word = corpora.Dictionary([doc.split() for doc in processed_docs])
corpus = [id2word.doc2bow(text.split()) for text in processed_docs]

lda_gensim_model = gensim.models.LdaModel(corpus, num_topics=3, id2word=id2word, passes=10)
topics = lda_gensim_model.print_topics(num_words=3)
print("Gensim LDA Topics:\n", topics)
```

```python
topics = lda_gensim_model.print_topics(num_words=5)
for idx, topic in topics:
    print(f"Topic #{idx}: {topic}")
```

```python
from collections import defaultdict

# Assign each document to its most dominant topic
document_topics = defaultdict(list)
for idx, doc in enumerate(corpus):
    topic_distribution = lda_gensim_model.get_document_topics(doc)
    # Find the topic with the highest probability for this document
    dominant_topic = max(topic_distribution, key=lambda x: x[1])[0]
    document_topics[dominant_topic].append(idx)

# Display sample documents for each topic
num_samples = 4  # Number of documents to display per topic
print("\nSample Documents for Each Topic Using Gensim LDA:")

for topic_id, doc_indices in document_topics.items():
    print(f"\nTopic {topic_id + 1}:")
    print(f"Number of documents assigned to this topic: {len(doc_indices)}")
    print(f"Displaying {min(num_samples, len(doc_indices))} sample documents:\n")

    for i in range(min(num_samples, len(doc_indices))):
        doc_index = doc_indices[i]
        print(f"Sample Document {i + 1} (Document ID: {doc_index}):")
        # Display the first 100 words from the document
        print(' '.join(processed_docs[doc_index].split()[:100]) + "...\n")
```

```python
import pyLDAvis
import pyLDAvis.gensim_models as gensimvis

# Visualize the topics
pyLDAvis.enable_notebook()
lda_display = gensimvis.prepare(lda_gensim_model, corpus, id2word, sort_topics=False)
pyLDAvis.display(lda_display)
```

## 5. Non-Negative Matrix Factorization (NMF)

```python
nmf_model = NMF(n_components=3, random_state=42)
nmf_topic_matrix = nmf_model.fit_transform(tfidf_matrix)

print("NMF Topic Matrix shape:", nmf_topic_matrix.shape)
```

```python
# Number of sample documents to display per topic
num_samples = 4

# Find the top documents for each topic
for topic_idx in range(nmf_model.n_components):
    print(f"\nDocuments most associated with Topic {topic_idx + 1}:")

    # Get the column corresponding to the current topic
    topic_column = nmf_topic_matrix[:, topic_idx]

    # Get indices of the documents sorted by their importance to the topic
    top_doc_indices = np.argsort(topic_column)[::-1][:num_samples]

    # Display sample documents
    for doc_num, doc_index in enumerate(top_doc_indices):
        print(f"\nSample document {doc_num + 1} (Document ID: {doc_index}):")
        # Display the first 100 words from the document
        print(''.join(processed_docs[doc_index][:100]) + "...")
```

##### Visualizing NMF Topics:

```python
plt.scatter(nmf_topic_matrix[:, 0], nmf_topic_matrix[:, 1], c=clusters)
plt.title('NMF Topic Plot (3 categories)')
plt.xlabel('Topic 1')
plt.ylabel('Topic 2')
plt.show()
```
